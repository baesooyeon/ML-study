# 잠재 요인 기반 협업 필터링
## 잠재 요인 협업 필터링의 개요

- 잠재 요인 협업 필터링은 사용자-아이템 평점 행렬(희소행렬) 속에 숨어 있는 잠재 요인을 추출해 추천 예측을 할 수 있게 하는 기법입니다.
- 대규모 다차원 행렬을 SVD와 같은 행렬 분해(Matrix Factorization) 기법으로 분해하는 과정에서 **잠재 요인을 추출**하는데, 이 잠재 요인을 기반으로 **사용자-아이템 평점 행렬을 재구성**하면서 추천을 구현합니다.

## 잠재 요인 협업 필터링의 이해
잠재 요인 협업 필터링은 사용자-아이템 평점 행렬 속에 숨어 있는 잠재 요인을 추출해 추천 예측을 할 수 있게 하는 기법입니다.

## 행렬 분해를 통한 잠재 요인 협업 필터링


**행렬 분해의 목표**
- **희소 행렬 형태의 사용자-아이템 평점 행렬**을 밀집(Dense)행렬 형태의 **사용자-잠재 요인 행렬**과 **잠재 요인-아이템 행렬**로 분해한 뒤 이를 재결합하여 **밀집 행렬 형태의 사용자-아이템 평점 행렬을 생성** 하여 사용자에게 **새로운 아이템을 추천**하는 것입니다.

## 잠재 요인 기반의 행렬 분해 이해

- M : 총 사용자 수
- N : 총 아이템 수
- K : 잠재 요인의 차원 수
- R : MxN차원의 사용자-아이템 평점 행렬(희소행렬)
- P : 사용자와 잠재 요인과의 관계 값을 가지는 MxK 차원의 사용자-잠재 요인 행렬
- Q : 아이템과 잠재 요인과의 관계 값을 가지는 NxK 차원의 아이템-잠재 요인 행렬
- Q.T : Q매트릭스의 행과 열 값을 교환한 전치행렬

## 사용자-아이템 평점 행렬 분해 이슈
- SVD는 Missing Value가 없는 행렬에 적용 가능하며, 따라서 희소행렬 기반의 사용자-아이템 평점 행렬은 일반적인 SVD 방식으로는 분해할 수 없습니다.
- 경사 하강법(Gradient Descent)를 이용하여 P와 Q에 기반한 예측 R값이 실제 R값과 가장 최소의 오류를 가질 수 있도록 **반복적인 비용함수 최적화**를 통해 P와 Q를 최적화 유추합니다.

## 경사 하강법 기반의 행렬 분해 순서
1. P와 Q를 **임의의 값**을 가진 행렬로 설정합니다.
2. P와 Q.T가 값을 곱해 예측 R 행렬을 계산하고 예측 R행렬과 실제 R행렬에 해당하는 오류 값을 계산합니다.
3. 이 오류 값을 최소화 할 수 있도록 **P와 Q행렬을 적절한 값으로 각각 업데이트** 합니다.
4. **만족할 만한 오류 값을 가질때까지 2, 3번 작업을 반복**하면서 P와 Q 값을 업데이트해 근사화합니다.
